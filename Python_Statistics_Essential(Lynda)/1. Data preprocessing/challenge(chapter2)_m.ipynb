```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
df=pd.read_csv('tb.csv')
df.head()
```

```python
df.columns
```

```python
melt= df.melt(['country', 'year'],['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'], 'sexage', 'cases')
melt.head()
```

```python
melt['sex']= melt['sexage'].str.slice(0,1)
melt['age']=melt['sexage'].str.slice(1)
melt.head()
```

```python
for i, row in melt.iterrows():
    temp = row['age']
    if temp == '65':
        melt.at[i, 'age'] = '65+'
    elif len(temp) == 2:
        melt.at[i, 'age'] = temp[0:1] + '-' + temp[1:]
    elif len(temp) == 3:
        melt.at[i, 'age'] = temp[0:1] + '-' + temp[1:]
    elif len(temp) == 4:
        melt.at[i, 'age'] = temp[0:2] + '-' + temp[2:]
    elif temp == 'u':
        melt.at[i, 'age'] = np.nan
    else:
        print('Value in row {} was not changed'.format(i))
```

```python
melt.head()
```

```python
final = melt.dropna(subset=['cases'])
```

```python
final = final.drop('sexage', axis=1)
final.head()
```

```python
final.sort_values(['country','year','age','sex'])
final.head()
```

```python
final.to_csv('final_m.csv',index=False)
```
